```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit
import datetime as dt
```

```python
df1mg1 = pd.read_csv("results_1mgrun1.csv")
df1mg2 = pd.read_csv("results_1mgrun2.csv")
df1mg3 = pd.read_csv("results_1mgrun3.csv")
df1mg4 = pd.read_csv("results_1mgrun4.csv")
df1mg5 = pd.read_csv("results_1mgrun5.csv")

ad1mg = [df1mg1, df1mg2, df1mg3, df1mg4, df1mg5]
ad1mg = pd.concat(ad1mg, axis=0)
ad1mg['date'] = pd.to_datetime(ad1mg['time'])
ad1mg = ad1mg.sort_values('date')
ad1mg.tail()
```

```python

df5mg3 = pd.read_csv("results_5mgrun3.csv")
df5mg4 = pd.read_csv("results_5mgrun4.csv")
df5mg5 = pd.read_csv("results_5mgrun5.csv")

ad5mg = [df5mg3, df5mg4, df5mg5]
ad5mg = pd.concat(ad5mg, axis=0)
ad5mg['date'] = pd.to_datetime(ad5mg['time'])
ad5mg = ad5mg.sort_values('date')
ad5mg.tail()
```

```python
# df10mg1 = pd.read_csv("results_10mgrun1.csv")
df10mg2 = pd.read_csv("results_10mgrun2.csv")
df10mg3 = pd.read_csv("results_10mgrun3.csv")
df10mg4 = pd.read_csv("results_10mgrun4.csv")
df10mg10 = pd.read_csv("results_10mgrun5.csv")

ad10mg = [df10mg2, df10mg3, df10mg4, df10mg10]
ad10mg = pd.concat(ad10mg, axis=0)
ad10mg['date'] = pd.to_datetime(ad10mg['time'])
ad10mg = ad10mg.sort_values('date')
ad10mg.tail()
```

```python

fig = plt.figure(figsize=(10,5))

x = ad1mg['date']
y = ad1mg['Absorbance']
ax1 = fig.add_subplot(111)
ax1.plot(x, y, '.')
ax1.set_xlabel("date")
ax1.set_ylabel("Absorbance")
ax1.set_title("1 mg/L")
plt.grid()
plt.yticks(np.arange(0, 2, 0.3))



```

```python
fig = plt.figure(figsize=(10,10))

df5mg1 = pd.read_csv("results_5mgrun1.csv")
df5mg2 = pd.read_csv("results_5mgrun2.csv")

ad5mgFirst2 = [df5mg1,df5mg2]
ad5mgFirst2 = pd.concat(ad5mgFirst2, axis=0)
ad5mgFirst2['date'] = pd.to_datetime(ad5mgFirst2['time'])
ad5mgFirst2 = ad5mgFirst2.sort_values('date')

x = ad5mgFirst2['date']
y = ad5mgFirst2['Absorbance']
ax4 = fig.add_subplot(211)
ax4.plot(x, y, '.')
# ax4.set_xlabel("date")
ax4.set_ylabel("Absorbance")
ax4.set_title("5 mg/L")
ax4.grid()
plt.yticks(np.arange(0, 2, 0.3))


x = ad5mg['date']
y = ad5mg['Absorbance']
ax2 = fig.add_subplot(212)
ax2.plot(x, y, '.')
ax2.set_xlabel("date")
ax2.set_ylabel("Absorbance")
# ax2.set_title("5 mg/L")
plt.grid()
plt.yticks(np.arange(0, 2, 0.3))


```

```python
fig = plt.figure(figsize=(10,10))


df10mg1['date'] = pd.to_datetime(df10mg1['time'])
df10mg1 = df10mg1.sort_values('date')

x = df10mg1['date']
y = df10mg1['Absorbance']
ax4 = fig.add_subplot(211)
ax4.plot(x, y, '.')
# ax4.set_xlabel("date")
ax4.set_ylabel("Absorbance")
ax4.set_title("10 mg/L")
ax4.grid()
plt.yticks(np.arange(0, 2, 0.3))

x = ad10mg['date']
y = ad10mg['Absorbance']
ax3 = fig.add_subplot(212)
ax3.plot(x, y, '.')
ax3.set_xlabel("date")
ax3.set_ylabel("Absorbance")
# ax3.set_title("10 mg/L")
ax3.grid()
plt.yticks(np.arange(0, 2, 0.3))
```

```python
ad1mg['actualC'] = 1
ad5mg['actualC'] = 5
ad10mg['actualC'] = 10


frame = [ad1mg, ad5mg, ad10mg]
all_Data = pd.concat(frame)

x = all_Data['actualC']
y = all_Data["Absorbance"]
b, m = polyfit(x, y, 1)

fig = plt.figure(figsize=(10,5))
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.xlabel("concentraion")
plt.ylabel("absorbance")
plt.grid()
plt.yticks(np.arange(0, 2, 0.3))

plt.xticks(np.arange(x.min(), x.max(), 1))
```

```python
# Checking Intesnity vs Absorbance changes

test = pd.read_csv("results_1mgrun2_test.csv")

test['date'] = pd.to_datetime(test['time'])
test = test.sort_values('date')
fig = plt.figure(figsize=(10,10))

x = test['date']

y = test['Absorbance']
ax4 = fig.add_subplot(211)
ax4.plot(x, y, '-')
ax4.set_xlabel("date")
ax4.set_ylabel("Absorbance")
ax4.set_title("1 mg/L")
ax4.grid()
# plt.yticks(np.arange(0,1, 0.1))


y = test['Reference_Intensity']
ax4 = fig.add_subplot(212)
ax4.plot(x, y, '-')
ax4.set_xlabel("date")
ax4.set_ylabel("Reference_Intensity")
ax4.set_title("1 mg/L")
plt.yticks(np.arange(y.min(), y.max(), 5))

ax4.grid()
```

```python
# ########################################
# Slicing the time for 4-15 minutes data 
# ########################################

test_sliced_4_min_onward = test[8:30]

fig = plt.figure(figsize=(10,10))

x = test_sliced_4_min_onward['date']

y = test_sliced_4_min_onward['Absorbance']
ax4 = fig.add_subplot(211)
ax4.plot(x, y, '-')
ax4.set_xlabel("date")
ax4.set_ylabel("Absorbance")
ax4.set_title("1 mg/L")
ax4.grid()
# plt.yticks(np.arange(0, 2, 0.3))


y = test_sliced_4_min_onward['Reference_Intensity']
ax4 = fig.add_subplot(212)
ax4.plot(x, y, '-')
ax4.set_xlabel("date")
ax4.set_ylabel("Reference_Intensity")
ax4.set_title("1 mg/L")

ax4.grid()
```
