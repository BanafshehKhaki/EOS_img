```python
# Checking Intesnity vs Absorbance changes

test = pd.read_csv("results_1mgrun1.csv")

test['date'] = pd.to_datetime(test['time'])
test = test.sort_values('date')
fig = plt.figure(figsize=(10,10))

x = test['date']

y = test['Absorbance']
ax4 = fig.add_subplot(211)
ax4.plot(x, y, '-')
ax4.set_xlabel("date")
ax4.set_ylabel("Absorbance")
ax4.set_title("1 mg/L")
ax4.grid()
# plt.yticks(np.arange(0,1, 0.1))


y = test['Reference_Intensity']
ax4 = fig.add_subplot(212)
ax4.plot(x, y, '-')
ax4.set_xlabel("date")
ax4.set_ylabel("Reference_Intensity")
ax4.set_title("1 mg/L")
plt.yticks(np.arange(y.min(), y.max(), 5))

ax4.grid()
```
